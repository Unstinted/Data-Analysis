```python
# Visualization 1: Distribution of feedback by gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='Feedback', data=df)
plt.title('Distribution of Feedback by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Feedback')
plt.show()
```

```python
# Visualization 2: Relationship between monthly income and feedback
plt.figure(figsize=(12, 8))
sns.countplot(x='Monthly Income', hue='Feedback', data=df, order=['No Income', 'Below Rs.10000', 'Rs.10001 - Rs.25000', 'Rs.25001 - Rs.50000', 'Above Rs.50000'])
plt.title('Relationship between Monthly Income and Feedback')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Feedback')
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Geographic distribution of positive vs. negative feedback
positive_feedback = df[df['Feedback'] == 'Positive']
negative_feedback = df[df['Feedback'].str.contains('Negative', na=False)]

plt.figure(figsize=(10, 6))
plt.scatter(positive_feedback['longitude'], positive_feedback['latitude'], color='green', alpha=0.5, label='Positive Feedback')
plt.scatter(negative_feedback['longitude'], negative_feedback['latitude'], color='red', alpha=0.5, label='Negative Feedback')
plt.title('Geographic Distribution of Feedback')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```

```python
import folium

# Create a map centered at an average location
average_lat = df['latitude'].mean()
average_lon = df['longitude'].mean()
feedback_map = folium.Map(location=[average_lat, average_lon], zoom_start=12)

# Add points to the map
for idx, row in df.iterrows():
    if row['Feedback'] == 'Positive':
        color = 'green'
    else:
        color = 'red'
    folium.CircleMarker(location=[row['latitude'], row['longitude']],
                        radius=5,
                        color=color,
                        fill=True,
                        fill_color=color,
                        fill_opacity=0.6
                       ).add_to(feedback_map)

# Display the map
feedback_map
```

```python
feedback_map.save('feedback_map.html')
```

```python
# Visualization: Age Distribution and Feedback
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Age', hue='Feedback', multiple='stack', bins=30, palette='viridis')
plt.title('Age Distribution and Feedback')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Feedback')
plt.show()
```

```python
# Visualization: Occupation and Feedback Analysis
plt.figure(figsize=(14, 8))
sns.countplot(y='Occupation', hue='Feedback', data=df, palette='coolwarm')
plt.title('Occupation and Feedback Analysis')
plt.xlabel('Count')
plt.ylabel('Occupation')
plt.legend(title='Feedback')
plt.tight_layout()  # Adjust layout to make room for the plot elements
plt.show()
```

```python
# Filter the data for students
students_data = df[df['Occupation'] == 'Student']

# Summary statistics for student demographics
student_age_distribution = students_data['Age'].describe()
student_monthly_income_distribution = students_data['Monthly Income'].value_counts()
student_educational_qualifications_distribution = students_data['Educational Qualifications'].value_counts()
student_feedback_distribution = students_data['Feedback'].value_counts()

print("Student Age Distribution:\n", student_age_distribution)
print("\nStudent Monthly Income Distribution:\n", student_monthly_income_distribution)
print("\nStudent Educational Qualifications Distribution:\n", student_educational_qualifications_distribution)
print("\nStudent Feedback Distribution:\n", student_feedback_distribution)
```

```python
# Plotting educational qualification by monthly income
plt.figure(figsize=(14, 8))
sns.countplot(x='Educational Qualifications', hue='Monthly Income', data=df,
              order=['Graduate', 'Post Graduate', 'Ph.D'],
              hue_order=['No Income', 'Below Rs.10000', '10001 to 25000', '25001 to 50000', 'More than 50000'],
              palette='Spectral')
plt.title('Educational Qualification by Monthly Income')
plt.xlabel('Educational Qualifications')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Monthly Income', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
